### Откроем датасет

Работать будем с датасетом 20newsgropus.<br/>
Он содержит письма. Чтобы не переобучаться на мусоре вроде хедеров и цитирований, удалим их с помощью параметра remove.

```python
from sklearn.datasets import fetch_20newsgroups
twenty = fetch_20newsgroups(subset='all', remove=('headers', 'footers', 'quotes'))

# Посмотрим на текст.
print(twenty.data[4])
```

### Выполним токенизацию с помощью nltk (см. семинар 1)

```python
from nltk import word_tokenize
from collections import Counter
import string

docs = []
for text in twenty.data:
    tokens = [w.lower() for w in word_tokenize(text) if not w in string.punctuation]
    docs.append(tokens)
```

```python
print('Токенизированный текст:')
print(docs[4])
```

### Удалим стоп-слова из текстов

Эти слова часто встречаются в текстах, вне зависимости от тематики. Поэтому в данной задаче они нам будут только мешаться.

```python
from nltk.corpus import stopwords
stop = set(stopwords.words('english'))
print(stop)
```

```python
new_docs = []
for tokens in docs:
    new_docs.append([token for token in tokens if not token in stop])
docs = new_docs
```

```python
print(docs[4])
```

### Построим словарь по корпусу

```python
from gensim import corpora

dictionary = corpora.Dictionary(docs)
print(dictionary)
```

```python
# Отфильтруем словарь
dictionary.filter_extremes(no_below=2, no_above=1, keep_n=300000)
print(dictionary)
```

```python
new_doc = "Hello world"
new_vec = dictionary.doc2bow(new_doc.lower().split())
```

```python
new_vec
```

```python
corpus = [dictionary.doc2bow(doc) for doc in docs]
```

```python
from gensim import models

tfidf = models.TfidfModel(corpus)
corpus_tfidf = tfidf[corpus]
```

### LSI

```python
lsi = models.LsiModel(corpus_tfidf, id2word=dictionary, num_topics=10)
```

```python
corpus_lsi = lsi[corpus_tfidf]
```

```python
lsi.print_topics(num_topics=10, num_words=20)
```

```python
lsi.show_topic(6, topn=30)
```

### LDA

```python
lda = models.LdaModel(corpus, id2word=dictionary, num_topics=20, passes=10)
```

```python
corpus_lda = lda[corpus]
```

```python
lda.print_topics(20)
```

```python
lda.show_topic(6, topn=30)
```

### Similarities

```python
from gensim import similarities
index = similarities.MatrixSimilarity(corpus_lsi)
```

```python
doc = docs[0]
vec_bow = dictionary.doc2bow(doc)
vec_lsi = lsi[tfidf[vec_bow]] # convert the query to LSI space
vec_lda = lda[vec_bow]
print(vec_lda)
```

```python
sims = index[vec_lsi] # perform a similarity query against the corpus
print(list(enumerate(sims)))
```
